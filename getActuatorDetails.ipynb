```python
import numpy as np
import time
import pandas as pd
#import ipywidgets as widgets
#from IPython.display import display
```

```python
start = time.process_time()
df = pd.read_csv('u20.csv', dtype='str')
df = df.drop(df.columns[0], axis = 1)
print('csv time taken:',time.process_time() - start)
```

```python
df
```

```python
list = []

for i in range(len(df)):
    if (df['TYPE'][i] == "COUTR" or df['TYPE'][i] == "ROUT") and "LPO" not in df['NAME'][-3:]:
        temp_list = []
        cp_name = df['NAME'][i].split('_')[1].split(':')[0]
        if 'LPO' not in cp_name:
            temp_list = [cp_name[:-1], df['DESCRP'][i] ]
            list.append(temp_list)


finalFrame = pd.DataFrame(list, columns = ['CP', 'Description'])
```

```python
finalFrame.to_csv('Actuators.csv')
```
